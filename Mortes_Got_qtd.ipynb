```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
got = pd.read_csv('game_of_thrones_character_deaths.csv')
```

```python
# Somar as mortes por livro
mortes_por_livro = got.groupby('book_of_death').size()

# Criar o gráfico de pizza
plt.pie(mortes_por_livro, labels=mortes_por_livro.index, autopct='%1.1f%%')

total_mortes = sum(mortes_por_livro)  # Calcula o total de mortes
plt.gca().annotate(f'Total: {total_mortes}', xy=(0.5, 0.5), fontsize=12, ha='center')

plt.title('Mortes em Game of Thrones')
plt.legend()
plt.show()
```
